```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1 = np.linspace(0,2*np.pi,500)
y_1 = np.linspace(0,2*np.pi,500)
```

```python
plt.scatter(np.cos(x_1), np.sin(y_1))
```

```python
mu, sigma = 0, 1


x1_noise = np.random.normal(mu, sigma,500)
y1_noise = np.random.normal(mu, sigma,500)
```

```python
x1 = 10*np.cos(x_1) + 3*np.cos(x1_noise)
y1 = 10*np.sin(y_1) + 3*np.sin(y1_noise)
```

```python
df = pd.DataFrame({'x': x1, 'y': y1})
df['z'] = np.random.uniform(-1, 1, 500)
```

```python
sns.scatterplot(data=df, x="x", y="y")
```

```python
sns.set_style ("darkgrid")

seaborn_plot = plt.axes (projection='3d')
seaborn_plot.scatter3D (df['z'], df['x'], df['y'])
seaborn_plot.set_xlabel ('x')
seaborn_plot.set_ylabel ('y')
seaborn_plot.set_zlabel ('z')
# plt.axis('equal')
plt.show ()

# this plot doesn't render properly because the aspect rations of the three axis are normalised and not proprtional - look into thus
```

```python
def FUNC_Z(x_plot, y_plot):
    return 50 - (x_plot**2 + y_plot**2)
sns.set_style('whitegrid')
num = 40
x_plot1 = np.linspace(-4, 4, num)
y_plot1 = np.linspace(-4, 4, num)
plot1, plot2 = np.meshgrid(x_plot1, y_plot1)
plot3 = FUNC_Z(plot1, plot2)
axes = plt.axes(projection='3d')
axes.plot_surface(plot1, plot2, plot3)
plt.show ()
```

```python
fig = plt.figure()

ax = plt.axes(projection='3d')
zline = np.linspace(0, 20, 2000)
xline = np.sin(zline)
yline = np.cos(zline)
xcoff = np.linspace(0, 1, 2000)
ycoff = np.linspace(0, 1, 2000)
ax.plot3D(xcoff * xline, ycoff * yline, zline, 'gray')

coff2 = np.linspace(0, 1, 1000)
std = np.linspace(0, .2, 1000)
zdata = np.linspace(0, 20, 1000)
xdata = coff2*np.sin(zdata) + std * np.random.randn(1000)
ydata = coff2*np.cos(zdata) + std * np.random.randn(1000)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='cool');

plt.show()
```
